## 1. セキュリティの3要素：CIA トライアド

情報セキュリティの目標は3つの言葉で表現できます。

| 要素 | 英語 | 意味 | 破られた例 |
|------|------|------|----------|
| **機密性** | Confidentiality | 見せてはいけない人に情報が見えない | パスワードが漏洩する |
| **完全性** | Integrity | データが改ざんされていない | 送金額が書き換えられる |
| **可用性** | Availability | 必要なときにシステムが使える | DDoS攻撃でサービス停止 |

頭文字を取って **CIA**（スパイ組織と同じ字ですが別物です）と呼びます。

### 日常の例で考える

あなたが銀行のオンラインシステムを使うとき：
- 自分の残高が他人に見えないこと → **機密性**
- 送金金額が途中で変わらないこと → **完全性**
- いつでもログインできること → **可用性**

攻撃者はこの3つのどれかを破ろうとします。

## 2. 脅威・脆弱性・リスク の違い

この3つの用語は混同されがちですが、明確に違います。

```
脆弱性（Vulnerability）  =  弱点（鍵のかかっていないドア）
脅威（Threat）           =  悪意ある行為者（泥棒）
リスク（Risk）           =  脆弱性 × 脅威 の組み合わせ（泥棒に入られる可能性）

リスク = 脆弱性が悪用された場合の影響 × その確率
```

### 例
- **脆弱性**: パスワードが短い（4桁）
- **脅威**: ブルートフォース攻撃（全パターン試す）
- **リスク**: アカウントを乗っ取られる

セキュリティ対策は「脆弱性をなくす」「脅威を排除する」「リスクを受容する」のいずれかです。

````python
# ブルートフォース攻撃のシミュレーション（教育用）
# 4桁PINが何秒で破れるかを計算する

import time

def simulate_brute_force(target_pin: str, delay_per_attempt: float = 0.0):
    """
    PINを総当たりで探す。
    delay_per_attempt: 1回の試行にかかる時間（秒）
    """
    attempts = 0
    start = time.time()
    
    for i in range(10000):  # 0000〜9999
        guess = f"{i:04d}"
        attempts += 1
        if guess == target_pin:
            elapsed = time.time() - start
            return attempts, elapsed
    
    return attempts, time.time() - start

# 4桁PIN「1234」を破る
attempts, elapsed = simulate_brute_force("1234")
print(f"4桁PIN '1234' を破るのに: {attempts} 回の試行")
print(f"コンピューターなら: {elapsed*1000:.2f} ミリ秒")
print(f"1秒に1回しか試せない場合: {attempts} 秒 ≈ {attempts/60:.1f} 分")
print()

# パスワードの長さと解読時間の関係
print("パスワード強度の比較:")
print(f"{'種類':<20} {'パターン数':<15} {'1000試行/秒で':>15}")
print("-" * 55)

cases = [
    ("4桁PIN", 10**4),
    ("6桁PIN", 10**6),
    ("8文字英数字", 62**8),
    ("12文字英数字+記号", 94**12),
]

for label, patterns in cases:
    seconds = patterns / 1000
    if seconds < 60:
        time_str = f"{seconds:.0f} 秒"
    elif seconds < 3600:
        time_str = f"{seconds/60:.0f} 分"
    elif seconds < 86400:
        time_str = f"{seconds/3600:.0f} 時間"
    elif seconds < 86400*365:
        time_str = f"{seconds/86400:.0f} 日"
    else:
        time_str = f"{seconds/86400/365:.2e} 年"
    print(f"{label:<20} {patterns:<15,} {time_str:>15}")
````
