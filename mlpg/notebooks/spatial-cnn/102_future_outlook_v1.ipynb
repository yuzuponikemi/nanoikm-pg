## 1. 深層学習の進化

```python
def visualize_evolution_timeline():
    """深層学習の進化のタイムライン"""
    fig, ax = plt.subplots(figsize=(18, 8))
    
    timeline = [
        (2012, 'AlexNet', 'CNN復活', 'lightblue'),
        (2014, 'VGGNet', '深いCNN', 'lightblue'),
        (2015, 'ResNet', 'スキップ接続', 'lightgreen'),
        (2015, 'U-Net', 'セグメンテーション', 'lightgreen'),
        (2017, 'Transformer', '自然言語処理', 'lightyellow'),
        (2020, 'ViT', '画像にTransformer', 'lightyellow'),
        (2020, 'NeRF', '3D表現', 'lightcoral'),
        (2023, '3DGS', '高速3D', 'lightcoral'),
    ]
    
    for i, (year, name, desc, color) in enumerate(timeline):
        x = (year - 2012) / 12
        y = 0.5 + 0.2 * ((-1) ** i)
        
        rect = plt.Rectangle((x, y - 0.08), 0.08, 0.16, 
                            facecolor=color, edgecolor='gray', linewidth=2)
        ax.add_patch(rect)
        ax.text(x + 0.04, y + 0.12, str(year), ha='center', fontsize=9)
        ax.text(x + 0.04, y, name, ha='center', va='center', fontsize=10, fontweight='bold')
        ax.text(x + 0.04, y - 0.12, desc, ha='center', fontsize=8)
    
    # 時間軸
    ax.axhline(y=0.5, color='black', linewidth=1, linestyle='-')
    
    # 凡例
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='lightblue', edgecolor='gray', label='CNN Era'),
        Patch(facecolor='lightgreen', edgecolor='gray', label='Architecture Innovation'),
        Patch(facecolor='lightyellow', edgecolor='gray', label='Transformer Era'),
        Patch(facecolor='lightcoral', edgecolor='gray', label='3D Understanding'),
    ]
    ax.legend(handles=legend_elements, loc='upper left', fontsize=10)
    
    ax.set_xlim(-0.05, 1)
    ax.set_ylim(0.2, 0.85)
    ax.axis('off')
    ax.set_title('Evolution of Visual Understanding through Deep Learning', fontsize=16, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

visualize_evolution_timeline()
```

```python
def explain_paradigm_shifts():
    """パラダイムシフトの説明"""
    print("="*70)
    print("深層学習のパラダイムシフト")
    print("="*70)
    
    print("""
【第1の波：CNN (2012-2017)】
  
  キーアイデア: 局所性 + 重み共有
  
  ・AlexNet (2012): GPU + ReLU + Dropoutで画像認識を革新
  ・VGGNet (2014): 「深さが重要」を証明（3×3のみで16-19層）
  ・GoogLeNet (2014): Inception module（マルチスケール）
  ・ResNet (2015): スキップ接続で100層以上が可能に
  
  限界: 長距離依存、大域的コンテキストの捕捉が困難

【第2の波：Transformer (2017-現在)】
  
  キーアイデア: 自己注意機構（全位置間の関係をモデル化）
  
  ・Transformer (2017): 「Attention Is All You Need」
  ・BERT/GPT (2018-19): NLPの事前学習革命
  ・ViT (2020): 画像をパッチに分割してTransformer
  ・Swin Transformer (2021): 階層的なWindow Attention
  
  トレードオフ: 計算量O(N²)、大量データが必要

【現在：ハイブリッドと3D】
  
  ・ConvNeXt: CNNをTransformer風に再設計
  ・NeRF/3DGS: 2D画像から3D理解へ
  ・Diffusion Models: 生成モデルの新時代
    """)

explain_paradigm_shifts()
```

```python
def compare_inductive_biases():
    """帰納バイアスの比較"""
    print("="*70)
    print("異なるアーキテクチャの帰納バイアス")
    print("="*70)
    
    comparison = [
        ('', 'CNN', 'Transformer', 'MLP-Mixer'),
        ('-'*12, '-'*18, '-'*18, '-'*18),
        ('局所性', '強い', 'なし', 'なし'),
        ('並進等変性', 'あり', 'なし', 'なし'),
        ('重み共有', 'あり（空間）', 'あり（位置）', 'あり（チャンネル）'),
        ('長距離依存', '弱い', '強い', '中程度'),
        ('パラメータ効率', '高い', '低い', '中程度'),
        ('必要データ量', '少ない', '多い', '多い'),
        ('計算量', 'O(HW)', 'O((HW)²)', 'O(HW)'),
    ]
    
    for row in comparison:
        print(f"{row[0]:<12} {row[1]:<18} {row[2]:<18} {row[3]:<18}")
    
    print("""
【洞察】
  ・CNNの帰納バイアスは「少ないデータでの学習」に有利
  ・Transformerは「大量データがあれば」より汎用的
  ・現代のベストプラクティスはハイブリッド（CNN + Attention）
    """)

compare_inductive_biases()
```

## 2. 3D理解への発展

```python
def visualize_2d_to_3d():
    """2Dから3Dへの発展を可視化"""
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    # 2D理解
    ax = axes[0]
    ax.text(0.5, 0.85, '2D Understanding', ha='center', fontsize=14, fontweight='bold')
    ax.text(0.5, 0.6, 'Image Classification\nSegmentation\nObject Detection', ha='center', fontsize=11)
    ax.text(0.5, 0.25, 'Input: Single Image\nOutput: Label/Mask', ha='center', fontsize=10)
    ax.set_facecolor('lightblue')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    
    # 2.5D理解
    ax = axes[1]
    ax.text(0.5, 0.85, '2.5D Understanding', ha='center', fontsize=14, fontweight='bold')
    ax.text(0.5, 0.6, 'Monocular Depth Estimation\nStereo Matching\nSurface Normal Estimation', ha='center', fontsize=11)
    ax.text(0.5, 0.25, 'Input: Image(s)\nOutput: Depth Map', ha='center', fontsize=10)
    ax.set_facecolor('lightgreen')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    
    # 3D理解
    ax = axes[2]
    ax.text(0.5, 0.85, '3D Understanding', ha='center', fontsize=14, fontweight='bold')
    ax.text(0.5, 0.6, 'NeRF\n3D Gaussian Splatting\nMesh Reconstruction', ha='center', fontsize=11)
    ax.text(0.5, 0.25, 'Input: Multi-View Images\nOutput: 3D Scene Representation', ha='center', fontsize=10)
    ax.set_facecolor('lightcoral')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    
    # 矢印
    for i in range(2):
        fig.patches.append(plt.FancyArrowPatch(
            (0.34 + i * 0.33, 0.5), (0.37 + i * 0.33, 0.5),
            transform=fig.transFigure,
            arrowstyle='->', mutation_scale=20, color='black'
        ))
    
    plt.suptitle('Evolution from 2D to 3D', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

visualize_2d_to_3d()
```

```python
def explain_3d_representations():
    """3D表現の種類を説明"""
    print("="*70)
    print("3Dシーン表現の種類")
    print("="*70)
    
    print("""
【明示的表現】
  
  1. 点群 (Point Cloud)
     ・3D点の集合
     ・メモリ効率が良い
     ・表面の連続性がない
  
  2. メッシュ (Mesh)
     ・頂点 + 面（三角形）
     ・レンダリングが高速
     ・トポロジーの変化が困難
  
  3. ボクセル (Voxel)
     ・3Dグリッド
     ・CNNを直接適用可能
     ・メモリ消費が膨大（O(N³)）

【暗黙的表現】
  
  4. NeRF (Neural Radiance Field)
     ・ニューラルネットで空間をパラメータ化
     ・任意の解像度で表現可能
     ・レンダリングが遅い
  
  5. 3D Gaussian Splatting
     ・3Dガウシアンの集合
     ・高速なラスタライズ
     ・NeRF品質 + リアルタイムレンダリング

【CNNとの接続】
  ・点群: PointNet（点ごとのMLP + 対称関数）
  ・ボクセル: 3D CNN（カーネルが3次元）
  ・NeRF/3DGS: CNN特徴を利用した初期化・条件付け
    """)

explain_3d_representations()
```

## 3. NeRFと3D Gaussian Splatting

```python
def visualize_nerf_concept():
    """NeRFの概念を可視化"""
    fig, ax = plt.subplots(figsize=(16, 8))
    
    # 入力
    ax.text(0.1, 0.7, 'Input', ha='center', fontsize=12, fontweight='bold')
    ax.text(0.1, 0.5, '(x, y, z, θ, φ)\n\nPosition + View Direction', ha='center', fontsize=10)
    rect = plt.Rectangle((0.02, 0.35), 0.16, 0.4, facecolor='lightblue', edgecolor='blue', linewidth=2)
    ax.add_patch(rect)
    
    # MLP
    ax.text(0.4, 0.7, 'MLP', ha='center', fontsize=12, fontweight='bold')
    ax.text(0.4, 0.5, 'F: R⁵ → R⁴\n\nNeural Network', ha='center', fontsize=10)
    rect = plt.Rectangle((0.3, 0.35), 0.2, 0.4, facecolor='lightgreen', edgecolor='green', linewidth=2)
    ax.add_patch(rect)
    
    # 出力
    ax.text(0.7, 0.7, 'Output', ha='center', fontsize=12, fontweight='bold')
    ax.text(0.7, 0.5, '(R, G, B, σ)\n\nColor + Density', ha='center', fontsize=10)
    rect = plt.Rectangle((0.6, 0.35), 0.2, 0.4, facecolor='lightcoral', edgecolor='red', linewidth=2)
    ax.add_patch(rect)
    
    # レンダリング
    ax.text(0.9, 0.55, 'Volume\nRendering', ha='center', fontsize=10)
    
    # 矢印
    ax.annotate('', xy=(0.3, 0.55), xytext=(0.18, 0.55),
               arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    ax.annotate('', xy=(0.6, 0.55), xytext=(0.5, 0.55),
               arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    ax.annotate('', xy=(0.85, 0.55), xytext=(0.8, 0.55),
               arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    
    # 式
    ax.text(0.5, 0.15, r'$C(r) = \int_{t_n}^{t_f} T(t) \sigma(r(t)) c(r(t), d) dt$',
           ha='center', fontsize=14)
    ax.text(0.5, 0.05, '(Integrate color and density along the ray)', ha='center', fontsize=10)
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 0.85)
    ax.axis('off')
    ax.set_title('NeRF (Neural Radiance Field) Concept', fontsize=16, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

visualize_nerf_concept()
```

```python
def visualize_3dgs_concept():
    """3D Gaussian Splattingの概念を可視化"""
    fig, ax = plt.subplots(figsize=(16, 8))
    
    # ガウシアンの描画
    np.random.seed(42)
    n_gaussians = 15
    
    for i in range(n_gaussians):
        x = 0.15 + np.random.rand() * 0.25
        y = 0.3 + np.random.rand() * 0.5
        size = 0.02 + np.random.rand() * 0.04
        color = plt.cm.viridis(np.random.rand())
        
        circle = plt.Circle((x, y), size, facecolor=color, alpha=0.6, edgecolor='gray')
        ax.add_patch(circle)
    
    ax.text(0.25, 0.85, '3D Gaussians', ha='center', fontsize=12, fontweight='bold')
    
    # パラメータ
    ax.text(0.55, 0.8, 'Parameters per Gaussian:', ha='left', fontsize=11, fontweight='bold')
    params = [
        '• 位置 (μ): x, y, z',
        '• 共分散 (Σ): 3×3行列',
        '• 不透明度 (α): スカラー',
        '• 色 (SH係数): 球面調和関数',
    ]
    for i, param in enumerate(params):
        ax.text(0.55, 0.7 - i*0.1, param, ha='left', fontsize=10)
    
    # レンダリングプロセス
    ax.text(0.55, 0.3, 'Rendering:', ha='left', fontsize=11, fontweight='bold')
    ax.text(0.55, 0.2, '1. Project Gaussians to 2D\n2. Tile-Based Sorting\n3. Alpha Blending', 
           ha='left', fontsize=10)
    
    # CNNとの接続
    ax.text(0.1, 0.1, '[Connection to Unit 0.4]\nReceptive Field ≈ Gaussian Spread\nLocality ≈ Finite Support Splat', 
           ha='left', fontsize=10, color='red',
           bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 0.95)
    ax.axis('off')
    ax.set_title('3D Gaussian Splatting (3DGS) Concept', fontsize=16, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

visualize_3dgs_concept()
```

```python
def compare_nerf_3dgs():
    """NeRFと3DGSの比較"""
    print("="*70)
    print("NeRF vs 3D Gaussian Splatting")
    print("="*70)
    
    comparison = [
        ('', 'NeRF', '3DGS'),
        ('-'*15, '-'*25, '-'*25),
        ('表現', '連続（暗黙的）', '離散（明示的）'),
        ('レンダリング', 'ボリュームレンダリング', 'ラスタライズ'),
        ('速度', '遅い（秒〜分/フレーム）', '速い（100+ fps）'),
        ('学習', 'MLPを最適化', 'ガウシアンを最適化'),
        ('メモリ', '少ない', '多い（ガウシアン数）'),
        ('編集', '困難', '比較的容易'),
        ('品質', '高い', '同等〜高い'),
    ]
    
    for row in comparison:
        print(f"{row[0]:<15} {row[1]:<25} {row[2]:<25}")
    
    print("""
【共通点】
  ・マルチビュー画像から3Dシーンを学習
  ・微分可能レンダリングで最適化
  ・新視点合成（Novel View Synthesis）が可能

【CNNからの知識の活用】
  ・特徴抽出: CNN特徴で初期化や条件付け
  ・深度推定: CNN深度で点群を初期化
  ・セグメンテーション: 3Dでのオブジェクト分離に活用
    """)

compare_nerf_3dgs()
```

## 4. NeoVerseへの道

```python
def visualize_neoverse_roadmap():
    """NeoVerseプロジェクトへのロードマップ"""
    fig, ax = plt.subplots(figsize=(16, 10))
    
    # 基礎層
    foundation = [
        (0.15, 0.85, 'Unit 0.4\n空間的帰納バイアス', 'lightblue'),
        (0.45, 0.85, '3D表現の基礎', 'lightblue'),
        (0.75, 0.85, '微分可能レンダリング', 'lightblue'),
    ]
    
    # 中間層
    intermediate = [
        (0.3, 0.55, 'NeRF理解', 'lightgreen'),
        (0.6, 0.55, '3DGS実装', 'lightgreen'),
    ]
    
    # 応用層
    application = [
        (0.45, 0.25, 'NeoVerse\n独自3Dシステム', 'gold'),
    ]
    
    for items, height in [(foundation, 0.15), (intermediate, 0.18), (application, 0.22)]:
        for x, y, text, color in items:
            rect = plt.Rectangle((x - 0.1, y - height/2), 0.2, height,
                                facecolor=color, edgecolor='gray', linewidth=2)
            ax.add_patch(rect)
            ax.text(x, y, text, ha='center', va='center', fontsize=10, fontweight='bold')
    
    # 矢印
    arrows = [
        ((0.15, 0.77), (0.3, 0.64)),
        ((0.45, 0.77), (0.3, 0.64)),
        ((0.45, 0.77), (0.6, 0.64)),
        ((0.75, 0.77), (0.6, 0.64)),
        ((0.3, 0.46), (0.45, 0.36)),
        ((0.6, 0.46), (0.45, 0.36)),
    ]
    
    for start, end in arrows:
        ax.annotate('', xy=end, xytext=start,
                   arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    
    # 現在地マーカー
    ax.plot(0.15, 0.85, 'r*', markersize=20)
    ax.text(0.15, 0.98, '← You are here!', ha='center', fontsize=12, color='red', fontweight='bold')
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1.05)
    ax.axis('off')
    ax.set_title('Roadmap to the NeoVerse Project', fontsize=16, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

visualize_neoverse_roadmap()
```

```python
def explain_knowledge_transfer():
    """Unit 0.4から3DGSへの知識転移"""
    print("="*70)
    print("Unit 0.4の知識がどう3DGSに活きるか")
    print("="*70)
    
    print("""
【畳み込みの理解 → スプラッティングの理解】
  
  CNNの畳み込み: 周囲を見て自分の値を決める
  3DGSのスプラット: 周囲に影響を与えるガウシアン
  
  → 「局所的な影響範囲」という共通概念

【受容野 → ガウシアンのスケール】
  
  CNNの受容野: 1ピクセルがどれだけの入力を見るか
  3DGSの共分散: ガウシアンがどれだけの空間をカバーするか
  
  → 「どの範囲の情報を統合するか」という設計思想

【帰納バイアス → 3D表現の設計】
  
  CNNの帰納バイアス: 「画像には局所的なパターンがある」
  3DGSの帰納バイアス: 「3Dシーンは局所的なガウシアンで表現できる」
  
  → 「事前知識を構造に埋め込む」という設計原理

【スキップ接続 → マルチスケール表現】
  
  U-Netのスキップ: 異なる解像度の情報を統合
  3DGSのマルチスケール: 異なるサイズのガウシアンで詳細と大域を表現
  
  → 「階層的な情報統合」という共通のアイデア
    """)

explain_knowledge_transfer()
```
